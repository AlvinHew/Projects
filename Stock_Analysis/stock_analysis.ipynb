```python
!pip install yfinance
import yfinance as yf 
import pandas as pd
```

```python
table = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")
snp_com_symb = table[0]["Symbol"].tolist()
len(snp_com_symb)
```

```python
snp_com_symb = snp_com_symb#[-10:]
snp_com_symb.append("SPY")
snp_com_symb[-5:]
```

```python
symbols = snp_com_symb
yf_period = "32y" #from 1990
yf_interval = "1d"
yf_returns = yf.download(
        tickers = symbols,       # tickers list or string as well
        period = yf_period,      # optional, default is '1mo'
        interval = yf_interval,  # fetch data by intervaal
        group_by = 'ticker',     # group by ticker
        auto_adjust = True,      # adjust all OHLC (open-high-low-close)
        prepost = False,#True,          # download market hours data
        threads = True,          # threads for mass downloading
        proxy = None)            # proxy
yf_returns.head(3)

"""
S&P500 have 2 Failed downloads:
- BF.B: No data found for this date range, symbol may be delisted
- BRK.B: No data found, symbol may be delisted
"""
```

```python
# yf_returns["Close"] is the same as yf_returns.loc[:,"Close"] #[index, column]
#yf_returns = pd.DataFrame(yf_returns["Close"]) #wrap in DataFrame to remain DataFrame format
yf_returns = yf_returns.iloc[:, yf_returns.columns.get_level_values(1)=="Close"]
#yf_returns = round(yf_returns.pct_change()*100, 2)
yf_returns.head(3)
```

```python
yf_returns.columns = yf_returns.columns.droplevel(1)
yf_returns.head(3)
```

```python
yf_returns = round(yf_returns.pct_change()*100, 2)
yf_returns.head(3)
```

```python
yf_divdend = pd.DataFrame()   # initialize dataframe
for symbol in symbols:
    x = pd.DataFrame(yf.Ticker(symbol).dividends)
    #Only one column, so needa wrap it in pd.DataFrame to make one column still DataFrame
    x = x.rename(columns={"Dividends":symbol}) #If dun rename, when concat will merge tgt
    yf_divdend = pd.concat([yf_divdend,x], axis=1) #axis = 0 is row, 1 is column
yf_divdend.head(3)
```

```python
yf_divdend = yf_divdend[yf_divdend.index >= yf_returns.index[0]] #match yf_returns's start date (index[0])
yf_divdend.head(3)
```

<span style="color:b"><font size="3">yf_returns = yf_returns + yf_divdend</font></span>

\begin{align*} 
\mathsf{\text{Percentage Return} =  \frac{price_{2}-price_{1}}{price_{1}}} \times  100 + dividends (\%)
\end{align*}

```python
yf_returns = yf_returns.add(yf_divdend, fill_value = 0) 
#fill value is done BEFORE addition, TSLA not IPO yet
yf_returns.head(3)
```

```python
yf_returns["MONTH"] = yf_returns.index.strftime("%Y-%m")  # YEAR-MONTH
yf_returns.tail(3)
```

```python
yf_returns = yf_returns.groupby('MONTH').sum() #groupby(): no actual computation until the transform_fn (e.g. .sum()) is applied
yf_returns.tail(3)
```

```python
yf_returns = yf_returns.loc[yf_returns["SPY"] < 0]
yf_returns.tail(3)
```

```python
#https://stackoverflow.com/questions/45383144/subtracting-many-columns-in-a-df-by-one-column-in-another-df
yf_returns = yf_returns.sub(yf_returns['SPY'], axis=0)
yf_returns.tail(3)
```

```python
mean = yf_returns.mean()
mean
```
